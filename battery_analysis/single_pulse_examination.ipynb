```python
from data_tools import DBClient
import numpy as np
from datetime import datetime, timezone
import matplotlib.pyplot as plt
```

```python
client = DBClient()
```

```python
start_time = datetime(2024, 7, 16, 18, 39, 0, tzinfo=timezone.utc)
end_time = datetime(2024, 7, 16, 18, 39, 40, tzinfo=timezone.utc)
```

```python
voltage_data = client.query_time_series(start_time, end_time, field="TotalPackVoltage", units="V", granularity=0.1)
current_data = client.query_time_series(start_time, end_time, field="PackCurrent", granularity=0.2, units="A")
```

```python
fig, ax1 = plt.subplots()
ax1.set_title("Multiple discharge pulse shape")

ax1.plot(voltage_data.datetime_x_axis, voltage_data, label="Pack Voltage", color="tab:red")
ax1.set_ylabel("V")
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(current_data.datetime_x_axis, current_data, label="Pack Current", color="tab:blue")
ax2.set_ylabel("A")
```
