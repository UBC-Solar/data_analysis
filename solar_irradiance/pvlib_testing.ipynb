```python
import pvlib
import pandas as pd
from data_tools import FSGPDayLaps
import matplotlib.pyplot as plt
```

```python
day_1_laps = FSGPDayLaps(1)
lap_start_times = []
for i in range(day_1_laps.get_lap_count()):
    lap_start_times.append(day_1_laps.get_start_utc(i + 1))

day_1_start_times = pd.DatetimeIndex(lap_start_times)

print(day_1_start_times)
```

```python
ncm_motorsports_park: pvlib.location.Location = pvlib.location.Location(
    latitude=37.00286254664568,
    longitude=-86.36698039236656,
)

cleasky_irradiances = ncm_motorsports_park.get_clearsky(day_1_start_times)
```

```python
plt.plot(day_1_start_times, cleasky_irradiances)
plt.show()
```
