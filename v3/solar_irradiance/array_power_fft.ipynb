```python
import numpy as np
import matplotlib.pyplot as plt
```

> Note: run `array_model_validation` to obtain the file `actual_array_power.npy`.

```python
actual_power = np.load('array_actual_power.npy', allow_pickle=True).item()
```

```python
def fourier_analysis(data: np.ndarray, title: str, period: float=0.1):
    plt.plot(data)
    plt.title(title)
    plt.xlabel("Time Since Start (s * 0.1)")
    plt.ylabel("W")
    plt.show()

    data_fft = np.fft.rfft(data)
    f_axis = np.fft.rfftfreq(data.size, period)

    plt.yscale("log")
    plt.plot(f_axis, np.absolute(data_fft))
    plt.title(f'{title} FFT magnitude, log scale')
    plt.xlabel("Frequency (Hz)")
    plt.show()

    plt.yscale("log")
    plt.plot(f_axis[:1000], np.absolute(data_fft)[:1000])
    plt.title(f'{title} FFT magnitude, log scale')
    plt.xlabel("Frequency (Hz)")
    plt.show()

    plt.yscale("log")
    plt.plot(f_axis[:100], np.absolute(data_fft)[:100])
    plt.title(f'{title} FFT magnitude, log scale')
    plt.xlabel("Frequency (Hz)")
    plt.show()
```

```python
fourier_analysis(actual_power['DAY_1']['COMBINED'], 'day 1 total array power', 0.1)
fourier_analysis(actual_power['DAY_3']['COMBINED'], 'day 3 total array power', 0.1)
fourier_analysis(np.array(actual_power['DAY_1']['COMBINED'])[25000:], 'day 1 total array power, removed start', 0.1)
fourier_analysis(np.array(actual_power['DAY_1']['COMBINED'])[:25000], 'day 1 total array power, only start', 0.1)
```

```python

day_1_data = np.array(actual_power['DAY_1']['COMBINED'])[25000:]
day_1_fft = np.fft.rfft(day_1_data)
f_axis_day_1 = np.fft.rfftfreq(day_1_data.size, 0.1)

day_3_data = actual_power['DAY_3']['COMBINED']
day_3_fft = np.fft.rfft(day_3_data)
f_axis_day_3 = np.fft.rfftfreq(day_3_data.size, 0.1)

plt.yscale("log")
plt.plot(f_axis_day_1, np.absolute(day_1_fft), alpha=0.7, label='day 1 array power, (start removed)')
plt.plot(f_axis_day_3, np.absolute(day_3_fft), alpha=0.7, label='day 3 array power')
plt.title(f'FFT magnitude, log scale')
plt.xlabel("Frequency (Hz)")
plt.legend(loc='best')
plt.show()

plt.yscale("log")
plt.plot(f_axis_day_1[:1000], np.absolute(day_1_fft)[:1000], alpha=0.7, label='day 1 array power, (start removed)')
plt.plot(f_axis_day_3[:1550], np.absolute(day_3_fft)[:1550], alpha=0.7, label='day 3 array power')
plt.title(f'FFT magnitude, log scale')
plt.xlabel("Frequency (Hz)")
plt.legend(loc='best')
plt.show()

plt.yscale("log")
plt.plot(f_axis_day_1[:100], np.absolute(day_1_fft)[:100], alpha=0.7, label='day 1 array power, (start removed)')
plt.plot(f_axis_day_3[:155], np.absolute(day_3_fft)[:155], alpha=0.7, label='day 3 array power')
plt.title(f'FFT magnitude, log scale')
plt.xlabel("Frequency (Hz)")
plt.legend(loc='best')
plt.show()
```
